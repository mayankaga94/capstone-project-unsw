```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('./goodbooks-10k/books.csv')
```

```python
books.head(110)
```

```python
book_tags = pd.read_csv('./goodbooks-10k/book_tags.csv')
```

```python
ratings = pd.read_csv('./goodbooks-10k/ratings.csv')
```

```python
tags = pd.read_csv('./goodbooks-10k/tags.csv')
```

```python
new_tags = pd.merge(book_tags, tags, how = 'left', on='tag_id')
```

```python
new_tags.head()
```

```python
def_tags = ['action', 'classics', 'comic', 'detective' , 'fantasy', 'historical', 'horror', 'literary', 
           'romance', 'sci-fi', 'biographies', 'cook', 'poetry','academic', 'business','art','cultural']
```

```python
len(def_tags)
```

```python
id_list = []
tag_list = []
for index, row in new_tags.iterrows():
    if row['goodreads_book_id'] in id_list:
        pass
    else:
        if row['tag_name'] in def_tags:
            id_list.append(row['goodreads_book_id'])
            tag_list.append(row['tag_name'])
```

```python
len(id_list)
```

```python
len(tag_list)
```

```python
to_export = pd.DataFrame({'goodreads_book_id':id_list,
                          'tag_name':tag_list})
```

```python
to_export.groupby(['tag_name']).count()
```

```python
to_export.to_csv(r'./consolidated dataset/consolidated_tags.csv', index=False)
```
